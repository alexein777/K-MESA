```python
import numpy as np
from clustering.kmesa import KMESA
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
X, y = make_blobs(n_samples=1000, centers=5, center_box=(-15, 15))
plt.scatter(X[:, 0], X[:, 1])
```

```python
est = KMESA(k_clusters=5,
             n_init=1,
             tol=1e-2,
             annealing_method='maxmin',
             annealing_prob_function='flex',
             alpha=0.3,
             annealing_weight_function='log',
             beta=5,
             convergence_tracking=True,
             annealing_tracking=True
             )
est.print_details()
est.plot_annealing_functions()
```

```python
est.fit(X)
est.print_clustering_info()
est.plot_clustered_data(X)
```

```python
est.plot_tracking_history(X)
```

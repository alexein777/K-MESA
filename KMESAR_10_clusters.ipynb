```python
import numpy as np
from clustering.kmesar import KMESAR
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
# X, y = make_blobs(n_samples=100000, centers=15, center_box=(-100, 100), cluster_std=4)
X, y = make_blobs(n_samples=1500, centers=10, center_box=(-20, 20))
```

```python
plt.scatter(X[:, 0], X[:, 1], s=4)
```

```python
est = KMESAR(k_clusters=10,
             init='k-means++',
             n_init=2,
             tol=1e-2,
             annealing_method='carousel',
             annealing_prob_function='sigmoid',
             alpha=20,
             annealing_weight_function='log',
             beta=4,
             convergence_tracking=True,
             annealing_tracking=True
             )
est.print_details()
est.plot_annealing_functions()
```

```python
est.fit(X)
est.print_clustering_info()
est.plot_clustered_data(X)
```

```python
est.plot_tracking_history(X, show_iter_mod=5, show_cc_labels=False, out_file='KMESAR_10_clusters')
```

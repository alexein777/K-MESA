```python
import numpy as np
from clustering.kmesar import KMESAR
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
X, y = make_blobs(n_samples=100000, centers=10, center_box=(-500, 500), cluster_std=30)
plt.scatter(X[:, 0], X[:, 1], s=4)
```

```python
est = KMESAR(k_clusters=10,
             init='random',
             n_init=10,
             tol=1e-3,
             ecr_method='random',
             annealing_method='carousel',
             annealing_prob_function='flex',
             alpha=0.2,
             annealing_weight_function='sigmoid',
             beta=50,
             convergence_tracking=True,
             annealing_tracking=True,
             ecr_tracking=True
             )
est.print_details()
est.plot_annealing_functions(n_iter=100)
```

```python
est.fit(X)
est.print_clustering_info()
```

```python
est.plot_clustered_data(X, show_cc_labels=False)
```

```python
est.plot_tracking_history(X, show_iter_mod=10, show_cc_labels=False, out_file='KMESAR_huge_dataset')
```

```python
import numpy as np
from clustering.kmesa import KMESA
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
X, y = make_blobs(n_samples=800, centers=20, center_box=(-120, 120), cluster_std=4)

fig = plt.figure(figsize=(5, 5))
plt.scatter(X[:, 0], X[:, 1], s=4)
fig.savefig('20_sparse')
```

```python
est = KMESA(k_clusters=20,
             init='random',
             n_init=5,
             tol=1e-3,
             ecr_method='random',
             annealing_method='max',
             annealing_prob_function='exp',
             alpha=100,
             annealing_weight_function='flex',
             beta=0.2,
             convergence_tracking=True,
             annealing_tracking=True,
             ecr_tracking=True
             )
est.print_details()
est.plot_annealing_functions(n_iter=100)
```

```python
est.fit(X)
est.print_clustering_info()
est.plot_clustered_data(X,
                        show_cc_labels=False,
                        out_file='KMESA_20_sparse_max_annealing')
```

```python
est.plot_iterations(X,
                    iterations=range(0, 16),
                    show_cc_labels=False,
                    out_file='KMESA_20_sparse_ECR_problem')


```

```python
import numpy as np
from clustering.kmesar import KMESA
from clustering.utils import load_dataset_3_clusters_separate
import matplotlib.pyplot as plt
```

```python
X = load_dataset_3_clusters_separate()
init_centroids = np.array([[-0.6, 0], [-0.2, 3], [0.4, 2]])
```

Random annealing application

```python
est = KMESA(k_clusters=3,
            n_init=1,
            tol=1e-2,
            init_centroids=init_centroids,
            annealing_method='random',
            annealing_prob_function='exp',
            alpha=15,
            annealing_weight_function='log',
            beta=5,
            convergence_tracking=True,
            annealing_tracking=True,
            ecr_tracking=True
            )
est.print_details()
est.plot_annealing_functions()
```

```python
est.fit(X)
est.print_clustering_info()
est.plot_clustered_data(X)
```

```python
est.plot_tracking_history(X, show_iter_mod=5, out_file='KMESA_simple_random_annealing')
```

```python
import numpy as np
from clustering.kmesar import KMESA
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
X, y = make_blobs(n_samples=1000, centers=20, center_box=(-120, 120), cluster_std=4)
plt.scatter(X[:, 0], X[:, 1], s=4)
```

```python
est = KMESA(k_clusters=20,
             init='random',
             n_init=3,
             tol=1e-3,
             ecr_method='random',
             annealing_method='centroid_split',
             annealing_prob_function='exp',
             alpha=50,
             annealing_weight_function='log',
             beta=7,
             convergence_tracking=True,
             annealing_tracking=True,
             ecr_tracking=True
             )
est.print_details()
est.plot_annealing_functions(n_iter=100)
```

```python
est.fit(X)
est.print_clustering_info()
```

```python
est.plot_clustered_data(X, show_cc_labels=False)
```

```python
est.plot_tracking_history(X, show_iter_mod=10, show_cc_labels=False, out_file='KMESA_20_clusters_ECR_problem')


```
